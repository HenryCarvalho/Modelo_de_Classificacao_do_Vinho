# üìã 7 - Detec√ß√£o de Outliers

Identificamos outliers em nossos dados e procederemos com sua remo√ß√£o.

```python
# Essa fun√ß√£o tem como objetivo detectar e retornar os √≠ndices dos outliers presentes no dataframe para as caracter√≠sticas especificadas.

def detect_outliers(df,features):
    outlier_indices = []
    
    for c in features:
        # 1st quartil
        Q1 = np.percentile(df[c],25)
        # 3st quartil
        Q3 = np.percentile(df[c],75)
        # IQR
        IQR = Q3 - Q1
        # Outlier Etapa
        outlier_step = IQR * 1.5
        # detectar outliers e seus √≠ndices
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # armazenar √≠ndices 
        outlier_indices.extend(outlier_list_col)
        
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 1.5) 
    
    return multiple_outliers
```

```python
# Imprimir o n√∫mero de outliers detectados em um conjunto de dados e, em seguida, retorna os registros que foram identificados como outliers.
print("n√∫mero de outliers detectados --> ",len(df_red.loc[detect_outliers(df_red,df_red.columns[:-1])]))
df_red.loc[detect_outliers(df_red,df_red.columns[:-1])]
```

```python
# Remover as linhas do dataframe que cont√™m outliers.
df_red =df_red.drop(detect_outliers(df_red,df_red.columns[:-1]),axis = 0).reset_index(drop = True)
```
